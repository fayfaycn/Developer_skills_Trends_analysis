## Extract information from the given web site
You will extract the data from the below web site: <br>


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
# Your code here
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
```

Download the webpage at the url


```python
#your code goes here
data  = requests.get(url).text
```

Create a soup object


```python
#your code goes here
soup = BeautifulSoup(data,"html5lib")
```

Scrape the `Language name` and `annual average salary`.


```python
#your code goes here
table = soup.find('table')

scraped_data = []

# Get all rows from the table
for row in table.find_all('tr'): # in html table row is represented by the tag
    # Get all columns in each row
    cols = row.find_all('td')
    # Skip header rows or empty rows
    if len(cols) > 0:
        # Skip the row that contains column headers as td elements (e.g., 'No.', 'Language')
        if cols[0].text.strip() == 'No.':
            continue
        language_name = cols[1].text.strip() # Programming language name is in the second column (index 1)
        annual_salary = cols[3].text.strip() # Average annual salary is in the fourth column (index 3)
        scraped_data.append([language_name, annual_salary])

print("Scraped Data:")
for item in scraped_data:
    print(f"Language: {item[0]}, Annual Salary: {item[1]}")
```

Save the scrapped data into a file named *popular-languages.csv*


```python
# your code goes here
import csv

# Define the CSV file name
csv_file = "popular-languages.csv"

# Define the header row
headers = ["Language", "Annual Average Salary"]

# Write data to CSV
with open(csv_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(headers)  # Write the header row
    writer.writerows(scraped_data) # Write the scraped data

print(f"Data successfully saved to {csv_file}")
```
